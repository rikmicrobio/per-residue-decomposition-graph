```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File path
file_path = r'C:\Users\rikmi\Desktop\graph_PRD\FINAL_DECOMP_MMPBSA_FILTERED.csv'

# Read CSV into DataFrame
df = pd.read_csv(file_path, delimiter='\t')

# Extracting required columns
df_extracted = df[['Resid 1', 'Resid 2', 'total energy', 'SD']]

# Determine SD ranges and corresponding sizes
sd_ranges = pd.cut(df_extracted['SD'], bins=[-np.inf, 1.4, 2, np.inf], precision=2, labels=['<1.4', '1.4-2', '>2'], include_lowest=True)
sizes = pd.Series(np.sqrt(df_extracted['SD'])) * 100  # Adjust scaling factor for increased differentiation

# Plotting
plt.figure(figsize=(12, 8))

# Create scatter plot with varying marker size based on SD
scatter = plt.scatter(df_extracted['Resid 1'], df_extracted['Resid 2'], 
                      s=sizes,  # Use scaled sizes based on sqrt(SD) with increased factor
                      c=df_extracted['total energy'], 
                      cmap='viridis', 
                      alpha=0.7)

# Create custom legend with marker sizes based on SD ranges
legend_sizes = [np.mean(sizes[sd_ranges == r]) for r in sd_ranges.unique()]  # Mean size for each SD range
legend_handles = []
for size, sd_range in zip(legend_sizes, sd_ranges.unique()):
    legend_handles.append(plt.scatter([], [], s=size, label=f'SD Range: {sd_range}', alpha=0.5))

# Add legend
plt.legend(handles=legend_handles, title='Standard Deviation (SD) Ranges', scatterpoints=1, labelspacing=1.2, loc='upper left')

# Add colorbar
cbar = plt.colorbar(scatter, label='Total Energy')
cbar.ax.set_ylabel('Total Energy', fontsize=14, fontweight='bold')  # Set label properties

# Customize plot labels and title
plt.xlabel('Residue 1', fontsize=14, fontweight='bold')
plt.ylabel('Residue 2', fontsize=14, fontweight='bold')
plt.title('Total Energy Interaction between Residues', fontsize=16, fontweight='bold')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.grid(True)
plt.tight_layout()
plt.show()
```
